## From this we decided to create a visualisation of this information on a graph. We did this using sns.lineplot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Here we convert the date column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'],format='%d/%m/%Y')

# Now we extract the month
df['Order Month'] = df['Order Date'].dt.month

# We group overall sales by Year and Month
df_Overall_Sales = df.groupby(['Order Year','Order Month']).agg({'Profit': 'sum', 'Sales': 'sum'}).reset_index()

# Define color palettes
sales_palette = sns.color_palette("Blues", n_colors=df_Overall_Sales['Order Year'].nunique())
profit_palette = sns.color_palette("Reds", n_colors=df_Overall_Sales['Order Year'].nunique())


# Plot
plt.figure(figsize=(12, 6))

sns.lineplot(data=df_Overall_Sales, x='Order Month', y='Sales', hue='Order Year', marker='o', palette=sales_palette, errorbar=None)
sns.lineplot(data=df_Overall_Sales, x='Order Month', y='Profit', hue='Order Year', marker='o', palette=profit_palette, errorbar=None)

plt.title('Overall Sales (Blue) and Profit (Orange) Across Years')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(df_Overall_Sales['Order Month'].unique())
plt.legend(title='Year')
plt.grid()
plt.show()
```

<span style="color: purple; font-weight: bold; font-size: 26px;">Key point #1:</span>
#### The graph reveals a clear seasonal pattern in overall sales, with peaks occurring consistently in the latter months of each year, suggesting predictable cycles in customer purchasing behavior. Despite these seasonal increases in sales, profit levels remain stable and significantly lower, indicating that profit margins are relatively low, likely due to high costs or promotional expenses associated with generating sales. The stagnation in profit, regardless of sales growth, suggests inconsistent or even rising cost levels as well as uneffective pricing strategies. This is because contrary to profit rising with sales volume as it should, it does not. Due to this, different marketing efforts, more effective pricing strategies, better cost control or advertising campaigns regarding seasonality could be implemented to boost profit.

## Overall Sales, Profit, Profit Ratio, Sales Share, Profit Share and Average Discount by Category and Year. Here we provide a breakdown of the key metrics by both category and year to provide a better understanding of what is actually occurring after a bit more of a deep dive. We performed an analysis of total sales by year which is then merged to our main Overall Sales Dataset. This is then sed to calculate Sales Share which allows us to see the proportion of each sales per category over each year. This is then plotted in a visual representation.

```python
# We group overall sales by Year and Sub-category
df_Overall_Sales = df.groupby(['Category','Order Year']).agg({'Profit': 'sum', 'Sales': 'sum', 'Discount': 'mean'}).reset_index()

# We calculate profit-to-sales ratio
df_Overall_Sales['ProfitRatio'] = df_Overall_Sales['Profit'] / df_Overall_Sales['Sales']

# Now we calculate total sales by Year for the sales share calculation
total_sales_by_year = df.groupby('Order Year')['Sales'].sum().reset_index()
total_sales_by_year.columns = ['Order Year', 'TotalSales']

df_Overall_Sales = df_Overall_Sales.merge(total_sales_by_year, on='Order Year')

# Sales share calculation
df_Overall_Sales['SalesShare'] = df_Overall_Sales['Sales'] / df_Overall_Sales['TotalSales']

# Calculate total profit by year for the profit share calculation
total_profit_by_year = df.groupby('Order Year')['Profit'].sum().reset_index()
total_profit_by_year.columns = ['Order Year', 'TotalProfit']


df_Overall_Sales = df_Overall_Sales.merge(total_profit_by_year, on='Order Year')

# Profit share calculation
df_Overall_Sales['ProfitShare'] = df_Overall_Sales['Profit'] / df_Overall_Sales['TotalProfit']

# Format
df_Overall_Sales['Profit'] = df_Overall_Sales['Profit'].apply(lambda x: f"{x:,.2f}")
df_Overall_Sales['Sales'] = df_Overall_Sales['Sales'].apply(lambda x: f"{x:,.2f}")
df_Overall_Sales['SalesShare'] = df_Overall_Sales['SalesShare'].apply(lambda x: f"{x:.2%}")  
df_Overall_Sales['ProfitShare'] = df_Overall_Sales['ProfitShare'].apply(lambda x: f"{x:.2%}") 
df_Overall_Sales['ProfitRatio'] = df_Overall_Sales['ProfitRatio'].apply(lambda x: f"{x:.2%}") 
df_Overall_Sales['Average Discount'] = df_Overall_Sales['Discount'].apply(lambda x: f"{x:.2%}") 

df_Overall_Sales[['Category', 'Order Year', 'Sales', 'Profit','ProfitRatio', 'SalesShare', 'ProfitShare', 'Average Discount']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot dataframe
if 'Profit' in df_Overall_Sales.columns:
    df_Overall_Sales['Profit'] = df_Overall_Sales['Profit'].replace({'$': '', ',': ''}, regex=True).astype(float)


sns.set(style='whitegrid')
plt.figure(figsize=(14, 8))
sns.barplot(data=df_Overall_Sales, x='Category', y='Profit', hue='Order Year', palette='rocket', alpha=0.8)

plt.title('Total Profit by Category and Order Year', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Profit ($)', fontsize=14)
plt.legend(title='Order Year')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

### Sales-Discount and Profit-Discount correlation by Sub-categories. Following from out detailed breakdown per Sub-category, we take a look at the link between sales discount and profit dicount by sub-category as we recognise from before that discount has a great potential link between why some high sales do not result in equally high profits. This is done using a for loop that provides a correlation co-efficient between the total discount and sales and profit.

```python
# Create a DataFrame to hold results
results = []
 
# Get all unique sub-categories
sub_categories = df['Sub-Category'].unique()
 
# Loop through each sub-category
for sub_category in sub_categories:
    # Filter for the specific sub-category
    sub_df = df[df['Sub-Category'] == sub_category].copy()
    # Check if the required columns exist in sub_df
    if 'Discount' not in sub_df.columns or 'Sales' not in sub_df.columns or 'Profit' not in sub_df.columns:
        print(f"Required columns missing in sub-category {sub_category}. Skipping...")
        continue
    # Add the hasDiscount and DiscountAmount columns
    sub_df = sub_df.assign(
        hasDiscount=(sub_df['Discount'] > 0).astype(int),
        DiscountAmount=sub_df['Sales'] * sub_df['Discount']
    )
    # Group by Product ID and sum the discounted amounts, discounts given, sales, and profit
    discounts_sub_df = sub_df.groupby('Product ID')[['hasDiscount', 'DiscountAmount', 'Sales', 'Profit']].sum().reset_index()
    # Rename the columns
    discounts_sub_df.columns = ['Product ID', 'DiscountsGiven', 'TotalDiscountAmount', 'TotalSales', 'TotalProfit']
    # Calculate correlations
    correlation_discounts_sales = discounts_sub_df[['TotalDiscountAmount', 'TotalSales']].corr().iloc[0, 1]
    correlation_discount_amount_profit = discounts_sub_df[['TotalDiscountAmount', 'TotalProfit']].corr().iloc[0, 1]
    # Append results
    results.append({
        'Sub-Category': sub_category,
        'Total Discounts Given': discounts_sub_df['DiscountsGiven'].sum(),
        'Total Discount Amount': discounts_sub_df['TotalDiscountAmount'].sum(),
        'Total Sales': discounts_sub_df['TotalSales'].sum(),
        'Total Profit': discounts_sub_df['TotalProfit'].sum(),
        'Correlation (Discount Amount vs Sales)': correlation_discounts_sales,
        'Correlation (Discount Amount vs Profit)': correlation_discount_amount_profit
    })
 
# Create a summary DataFrame
summary_df = pd.DataFrame(results)
 
# Display the summarized DataFrame
summary_df
```

#### The findings from this table reveal a wide range of correlations between discounts and profit across various sub-categories, underscoring the importance of a tailored discount strategy. Marketing specialists should carefully evaluate these correlations to ensure that discounts are aligned with profitability goals within each sub-category. For instance, in the Supplies sub-category, there is an almost perfectly negative correlation between discounts and profit, indicating that discounts here may reduce profitability rather than enhance it. This suggests that offering discounts on Supplies could be counterproductive. On the other hand, sub-categories like Copiers show a strong positive correlation between discounts and profit, where discounts appear to drive increased profitability. In such cases, discounts could be strategically leveraged to boost sales and overall profit. By considering the unique correlation of each sub-category, the marketing team can optimize discount strategies to support profitability more effectively.
